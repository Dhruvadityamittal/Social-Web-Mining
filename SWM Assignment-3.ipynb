```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.complete_graph(5)

nx.draw(G)
plt.show()
preds = nx.jaccard_coefficient(G, [(0, 1), (2, 3)])

print("Jaccard coefficient of the regular Graph")
for u, v, p in preds:
     print(f"({u}, {v}) -> {p:.8f}")
```

Given four persons A, B, C and D with likes on the following object IDs:
• likes(A) = [1,2,3]
• likes(B) = [2,3,4]
• likes(C) = [1, 6, 7]
• likes(D) = [4,5,6,7]
Calculate all pairwise similarities using the Jaccard similarity coefficient in a matrix. From
these values, draw a similarity graph with four nodes and the weighted edges derived from
the matrix.

```python
G.clear()
G.add_edges_from([("A",1),("A",2),("A",3)])
G.add_edges_from([("B",2),("B",3),("B",4)])
G.add_edges_from([("C",1),("C",2),("C",3)])
G.add_edges_from([("D",4),("D",5),("D",6),("D",7)])

nx.draw(G,with_labels = True)
plt.show()
```

```python
nx.nodes(G)
```

```python
preds =nx.jaccard_coefficient(G, [('A', 'B'),('A','C'),('A','D'),('B','C'),('B','D'),('C','D')])

for u, v, p in preds:
    print("Jaccards Similarity Between",f"({u}, {v}) -> {p:.8f}")
```
